**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["MARGEN_TOTAL_CALCULADO"] > 10000]
filtro6

#AQUI BUSCAMOS VER EL MARGEN POR ARRIBA DE 10000 PESOS QUE HA TENIDO LA EMPRESA Y VEMOS QUE SON SOLAMENTE 10 CASOS LOS CUALES SE TUVO ESTE MARGEN
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["PRECIO_UNITARIO"] > 750) & (df["FECHA_DOC"] == "2022-12-20")]
filtro7

#Cumplen las dos condiciones.

#AQUI BUSCAMOS QUE LA FECHA DEL 20 DE DICIEMBRE Y QUE EL PRECIO POR UNIDAD DEL ARTICULO VENDIDO HAYA SIDO POR ARRIBA DE 750 PESOS. VEMOS QUE SOLO 3 ARTICULOS FUERON VENDIDOS POR ARRIBA DE ESE PRECIO EN ESE DIA.
```

```python
#Filtro o
filtro8=df[(df["PRECIO_UNITARIO"] > 750) | (df["FECHA_DOC"] == "2022-12-20")]
filtro8

#AQUI HACEMOS ALGO SIMILAR A LO ANTERIOR PERO VENTAS CON PRECIO UNITARIO MAYOR A 750 Y QUE SEAN DE ESA FECHA VEMOS QUE SOM MAS DE 800 DATOS.
```

```python
#Filtro not
filtro9=df[~(df["PRECIO_UNITARIO"] > 750) & ~(df["FECHA_DOC"] == "2022-12-20")]
filtro9

#BUSCAMOS QUE EL PRECIO UNITARIO SEA MAYOR A 750 PERO NO ME CUENTEN LOS 3 LA FECHA ASIGNADA.
```
